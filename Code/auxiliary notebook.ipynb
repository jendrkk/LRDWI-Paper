```python
from bs4 import BeautifulSoup
import soupsieve as sv

html = """<html><body>
  <div class="target"><p>one</p></div>
  <div class="target"><p>two</p></div>
</body></html>"""

soup = BeautifulSoup(html, "html.parser")

# If you can provide a CSS selector, great:
css_selector = "div.target"    # equivalent for //div[@class='target']
nodes = sv.select(css_selector, soup)

print(nodes)

for n in nodes:
    # str(n) returns outer HTML of the tag
    print(str(n))
```
